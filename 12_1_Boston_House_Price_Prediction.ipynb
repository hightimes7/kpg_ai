### 보스턴 집값 예측 딥러닝

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split

import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
df = pd.read_csv("./dataset/housing.csv", delim_whitespace=True, header=None)
# delim_whitespace : 중간에 데이터가 공백으로 구분된 파일을 읽을 때 사용
df
```

```python
df_ko = pd.read_csv("./dataset/housing.csv" , delim_whitespace=True, 
                    names = ["범죄 발생 수", "주거 구역 비중", "상업 비중", "찰스강 위치", 
                             "일산화질소", "방 개수", "오래된 집 비율","고용시설 까지 거리", 
                             "순환고속도로 접근", "부동산 세율", "학생 및 교사 비율", 
                             "흑인 비율", "최저급여 인구 비율","집 가격"])
df_ko
```

```python
df_ko.info()
```

```python
dataset = df_ko.values
dataset
```

```python
np.shape(dataset)
```

```python
X = dataset[:,:13]
Y = dataset[:, 13]
```

```python
np.shape(X)
```

```python
np.shape(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
```

```python
X_train
```

```python
print('len(X) :', len(X))
print('len(X_train) :', len(X_train))
print('len(X_test) :', len(X_test))
```

```python
print('np.shape(X) : ', np.shape(X))
print('np.shape(X_train) :', np.shape(X_train))
print('np.shape(X_test) :', np.shape(X_test))
```

```python
print('len(Y) :', len(Y))
print('len(Y_train) :', len(Y_train))
print('len(Y_test) :', len(Y_test))
```

```python
print('np.shape(Y) : ', np.shape(Y))
print('np.shape(Y_train) :', np.shape(Y_train))
print('np.shape(Y_test) :', np.shape(Y_test))
```

```python
model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(X_train, Y_train, epochs=200, batch_size=10)
```

```python
Y_prediction = model.predict(X_test)
print(np.shape(Y_prediction))
print(Y_prediction)
```

```python
Y_prediction = model.predict(X_test).flatten()  # flatten() : shape를 1차원으로 변환
print(np.shape(Y_prediction))
print(Y_prediction)
```

```python
for i in range(10):
    label = Y_test[i]
    prediction = Y_prediction[i]
    print('실제가격 : {:.3f}, 예상가격 : {:.3f}'.format(label, prediction))
```
