### 폐암 수술 환자 생존율 예측 딥러닝
#### sklearn 이용한 학습 데이터와 테스트 데이터 분리 (7:3)

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
import tensorflow as tf
```

```python
np.random.seed(3)
tf.random.set_seed(3)
```

```python
Data_set = np.loadtxt('dataset/ThoraricSurgery.csv', delimiter=',')
Data_set
```

```python
X = Data_set[:,0:17]
Y = Data_set[:, 17]
```

```python
X
```

```python
Y
```

```python
print('X shape:',np.shape(X))
print('Y shape:',np.shape(Y))
```

```python
seed = 0
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state= seed)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu'))  # input_dim : X 데이터 17가지 노드 17개로 입력
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, Y_train, epochs=100, batch_size=10)
```

```python
model.evaluate(X_test, Y_test)  # 학습으로 구한 weight를 테스트에 적용
```

#### 학습 후 예측

```python
kim = np.array([[293,1,3.8,2.8,0,0,0,0,0,0,12,0,0,0,1,0,62]])
print(model.predict(kim))
```

```python
test_kim = model.predict(kim)*100
print('백분율로 표시 :', test_kim)
print('Kim 폐암 수술 후 생존율 예측 : %.2f' %test_kim, "%")
```

```python
park = np.array([[197,3,2.84,2.24,1,1,1,0,0,0,12,0,0,0,1,0,68]])
print(model.predict(park))
```

```python
test_park = model.predict(park)*100
print('백분율로 표시 :', test_park)
print('Park 폐암 수술 후 생존율 예측 : %.2f' %test_park, "%")
```
