### 아이리스 품종 분류 - TensorBoard

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
from datetime import datetime
import tensorboard
```

```python
df = pd.read_csv('dataset/iris.csv', names=['꽃받침길이','꽃받침넓이','꽃잎길이',
                                               '꽃잎넓이','종류'])
df
```

```python
dataset = df.values
type(dataset)
```

```python
X = dataset[:,0:4].astype(float)
Y_obj = dataset[:,4]
```

```python
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)
Y
```

```python
Y_encoded = tf.keras.utils.to_categorical(Y)
Y_encoded
```

```python
logdir = "log\\" + datetime.now().strftime("%Y%m%d-%H%M%S")
logdir
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir, histogram_freq=1)
```

```python
seed = 0
np.random.seed(3)
tf.random.set_seed(3)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_encoded, test_size=0.3, random_state=seed)
```

```python
model = Sequential()
model.add(Dense(16, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epochs=200,
          callbacks=[tensorboard_callback])
```
