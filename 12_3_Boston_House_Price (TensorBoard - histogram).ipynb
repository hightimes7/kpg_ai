### 보스턴 집값 예측 딥러닝

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split

import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
df_ko = pd.read_csv("./dataset/housing.csv" , delim_whitespace=True, 
                    names = ["범죄 발생 수", "주거 구역 비중", "상업 비중", "찰스강 위치", 
                             "일산화질소", "방 개수", "오래된 집 비율","고용시설 까지 거리", 
                             "순환고속도로 접근", "부동산 세율", "학생 및 교사 비율", 
                             "흑인 비율", "최저급여 인구 비율","집 가격"])
df_ko
```

```python
dataset = df_ko.values
dataset
```

```python
X = dataset[:,:13]
Y = dataset[:, 13]
```

```python
seed=0
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
```

```python
print("len(X) : ", len(X))
print("len(X_train) : ", len(X_train))
print("len(X_test) : ", len(X_test))
print("-"*50)
print("numpy.shape(X) : ", np.shape(X))
print("numpy.shape(X_train) : ", np.shape(X_train))
print("numpy.shape(X_test) : ", np.shape(X_test))
print("-"*50)
print("len(Y) : ", len(X))
print("len(Y_train) : ", len(X_train))
print("len(Y_test) : ", len(X_test))
print("-"*50)
print("numpy.shape(Y) : ", np.shape(Y))
print("numpy.shape(Y_train) : ", np.shape(Y_train))
print("numpy.shape(Y_test) : ", np.shape(Y_test))
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir, histogram_freq=1)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=200, batch_size=50, callbacks=[tensorboard_callback])
```
